#### 6. Create a new dataframe that only contains data from the *province* with the largest percentage changes in your chosen activity from *Question 5*, which you found in the previous question and name it as `df_province_largest`. Print the first 5 lines of `df_province_largest`


##### *********** Please Read the Following ***********
I was not sure whether you ask us to find the province with the largest percentage change **in the specific
date of question 5**, or **for the whole data set**; therefore, I decided to do the more challenging one and found the province with the largest perentage change in the whole dataset. 
Also, I should note that I am caculating the largest change based on the absolute values of changes. If we calculate the average without using the absolute values, we might be misled because there might be positive and negative values together.For instance, a +20 and a -20 change will neutralize eachother.

Also, I have changed the resampling type from **sum()** to **mean()** because I believe calcuating the average values of changes in a week better represent the weekly trends compared to when adding all changes together. For instance, if we sum the percentage changes for seven days of a week, we might reach to a change larger (smaller) than 100 (-100) percent. 

```python
# Please read the note above before reading the code. Thank you!

# Summarizing the dataset and calculating the absolute values of changes from the baseline
df_province_abs = df_province.iloc[:, [2,13]]
df_province_abs['workplaces_percent_change_from_baseline']=df_province_abs.iloc[:,1].abs()

# Finding the Province/State with largest changes from the baseline
df_groupby = df_province_abs.groupby('sub_region_1').mean()                                  
province_largest_index = df_groupby.workplaces_percent_change_from_baseline.idxmax()      # Finding the index of the province with the largest changes
df_province_largest = df_province.loc[df_province.sub_region_1 == province_largest_index] # Finding the corresponding value based on the found index

# Printing the first 5 rows of the DataFrame
df_province_largest.head(5)
```

```python
# Execute the cell with the following command to visualize the results
#Your chosen activity. eg: 'workplaces_percent_change_from_baseline'
activity = 'workplaces_percent_change_from_baseline' #YOUR CODE HERE 

df_province_largest['date'] = pd.to_datetime(df_province_largest['date']) # Changing the 'date' column into DateTime
                                                                          # So that it is recognizable by the resample function

df_province_largest.plot('date', activity)
df_province_largest.resample('7D', on='date').mean()[activity].plot()     
```
